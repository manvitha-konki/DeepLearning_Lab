```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.datasets import load_iris
```

```python
df1 = pd.read_csv(r"C:\Users\91898\Downloads\iris.csv")
df1
```

```python
df = df1[(df1["variety"] == 'Setosa') | (df1["variety"] == 'Versicolor')]
df['class'] = np.where(df["variety"] == 'Setosa', 0, 1)     # If variety is Setosa give '0' Else give '1'
df = df.drop(columns=['variety'])
df['bias']=1
```

```python
df
```

```python
# 80-20 SPLIT OF THE DATASET
dfTrain = pd.concat([df.iloc[0:40], df.iloc[50:90]]).reset_index(drop=True)
dfTest = pd.concat([df.iloc[40:50], df.iloc[90:100]]).reset_index(drop=True)
```

```python
a=np.array([0,0,0,0,0]) # INITIAL WEIGHT VECTOR
learningRate=0.0001 # LEARNING RATE
epochs=100 # MAX ITERATIONS
```

```python
#TRAINING THE PERCEPTRON
for it in range(epochs):
    mse=0
    for i in range (len(dfTrain)):
        W=np.array([dfTrain.loc[i,'sepal.length'],dfTrain.loc[i,'sepal.width'],dfTrain.loc[i,'petal.length'],dfTrain.loc[i,'petal.width'],dfTrain.loc[i,'bias']])
        if (a*W).sum() > 0:
            Output=1
        else:
            Output=0
        error=Output-dfTrain.loc[i,'class']
        a=a-learningRate*error*W
        mse+=error**2
    mse/=(len(dfTrain))
    print(f"Iteration {it+1}, MSE: {mse}")
    if mse==0: break

# FINAL WEIGHT VECTOR AFTER CONVERGENCE OF PERCEPTRON
print(f"The final Weight vector after convergence is {a}")

#TESTING THE PERCEPTRON
count=0
for i in range(len(dfTest)):
    W=np.array([dfTest.loc[i,'sepal.length'],dfTest.loc[i,'sepal.width'],dfTest.loc[i,'petal.length'],dfTest.loc[i,'petal.width'],dfTest.loc[i,'bias']])
    if (a*W).sum() > 0:
        Output=1
    else:
        Output=0
    if Output==dfTest.loc[i,'class']:
        count += 1

#ACCURACY
accuracy = count/len(dfTest) * 100
print(f"Accuracy on test set: {accuracy:.2f}%")

```
