```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
groundTruth = np.array([1,0])  # Actual Output - Ground Truth
Layers = np.array([3,3,3]) 
'''
The network has 3 layers:
Input layer with 3 neurons.
Hidden layer with 3 neurons.
Output layer with 3 neurons.
'''
learningRate = 0.5
```

```python
Input = np.array([1,0.7,1.2])
initailWeight = [None, np.array([[0.5,1.5,0.8], [0.8,0.2,-1.6]]), np.array([[0.9,-1.7,1.6],[1.2,2.1,-0.2]])]    # Weight Vectors
Output = np.zeros((len(Layers), max(Layers)))   # Output at each layer - len(layers) - No of layers and max(layers) - Highest No of neurons in any one of the layer
Error = np.zeros((len(Layers), max(Layers)))     # Error values at each layer initializing with zeros
```

```python
for it in range(1000):
    weightOld=initailWeight.copy()
    #FEED FORWARD
    for i in range(len(Layers)):    # Travel through Layers
        for j in range(Layers[i]):  # Travel through neurons
            if i == 0:  # Input Layer
                Output[i, j] = Input[j]
            else:   # Hidden or Output Layer
                if j == 0:  # If Bias Neurons, set output as '1'
                    Output[i, j] = 1
                else:   # Normal neurons
                    Output[i, j] = sigmoid(np.dot(initailWeight[i][j-1], Output[i-1, :Layers[i-1]]))
    
    Output=Output.T
    print(f"output differences are: {Output[1,2]-groundTruth[0]} and {Output[2,2]-groundTruth[1]}")
    '''
    This prints the difference between the outputs of certain neurons and the ground truth. It helps in tracking the model's progress.
    '''
    
    #BACK PROPAGATION
    #ERROR AT LAYER 2
    for j in range(1,3):
        Error[j, 2]=Output[j, 2]*(1-Output[j, 2])*(Output[j, 2]-groundTruth[j-1])
    
    #WEIGHT UPDATION AT LAYER 1->2
    for i in range(2):
        for j in range(3):
            initailWeight[2][i, j]=initailWeight[2][i, j]-learningRate*Error[i+1, 2]*Output[j, 1]
            
    #ERROR AT LAYER 1
    for i in range(1,3):
        alpha=0
        for j in range(0,2):
            alpha=alpha+Error[j+1, 2]*weightOld[2][j, i]
        Error[i, 1]=Output[i, 1]*(1-Output[i, 1])*alpha
        
    #WEIGHT UPDATION AT LAYER 0->1
    for i in range(2):
        for j in range(3):
            initailWeight[1][i, j]=initailWeight[1][i, j]-learningRate*Error[i+1, 1]*Output[j, 0]
```
